First, reading in 5 images with the help of the Pillow library. I choose to load the images via url, so first of all, the images need to be loaded.

````
# Read in images in NumPy array
from PIL import Image
import requests
from io import BytesIO
from numpy import asarray
import numpy as np

urls = [
        'https://64.media.tumblr.com/avatar_2232fd314ea0_128.pnj',
        'https://64.media.tumblr.com/avatar_a3ec0f8f9435_128.pnj',
        'https://i.pinimg.com/originals/b0/0a/f2/b00af2dea76f27983a94dab6cd7baab9.jpg',
        'https://64.media.tumblr.com/avatar_e9e7697832df_128.pnj',
        'https://64.media.tumblr.com/avatar_2232fd314ea0_128.pnj']
images = []
for url in urls:

  response = requests.get(url)
  img = Image.open(BytesIO(response.content))
  display(img)

  arr = asarray(img)
  print(arr)
  images.append(arr)
````

A get request is sent via the requests library, then the image is loaded from the response via BytesIO. After that, the image can be displayed. The array containing the image data is extracted with the help of numpy's asarray method. 

Then, the mean and the standard deviation needs to be calculated. Firstly, the image data needs to be striped to avoid the 0 standard deviation, originating from the 4th dimension of the image data, as we will only need the RGB channels. 

To calculate the values for the RGB channels properly, the calculations need to go first along the 0th and then along the 1th axis.

After that, to achieve a mean of 0 and a standard deviation of 1, the data needs to be standardized. For that the formula 

```
arr = (arr - mean) / std
```
is used.


````
for image in images:
  # Stripe data
  image = image[:, :, 0:3]

  # Calculate mean of the channels
  mean = np.mean(image, axis=(0, 1))

  # Calculate standard deviation of the channels
  std = np.std(image, axis=(0, 1))

  image = (image-mean) / std 

  
  print("Standardized mean: ", np.mean(image, axis=(0, 1)))

  print("Standardized standard deviation: ", np.std(image, axis=(0, 1)))
````

Then, a spectogram of a sound file needs to be displayed. For that purpose, I chose to work with wav files, as it seemed to be the easiest approach. 

First, the wave file is requested similarly as before. Then it is read with the help of the scipy library. The spectogram is generated with scipy's signal and is displayed via matplotlib.

````
from scipy import signal
from scipy.io import wavfile

response = requests.get('https://www2.cs.uic.edu/~i101/SoundFiles/gettysburg10.wav')

sample_rate, data = wavfile.read(BytesIO(response.content))

frequencies, times, spectrogram = signal.spectrogram(data, sample_rate)

plt.pcolormesh(times, frequencies, spectrogram)
plt.imshow(spectrogram, aspect='auto')
plt.ylabel('Frequency [Hz]')
plt.xlabel('Time [sec]')

plt.show()

````

Lastly, the content of the Deep learning wiki page needs to be loaded. First the url is opened with urlopen from the urllib.request library and is read as a html file.

````
from urllib.request import urlopen



page = urlopen('https://en.wikipedia.org/wiki/Deep_learning')
html = page.read()


````

To get the raw text from the html file, BeautifulSoup is used. 

````
from bs4 import BeautifulSoup


soup = BeautifulSoup(html)
raw_text = soup.get_text()

print(raw_text)

````

Having read the raw text, I chose to store the keys and values in a dictionary. To get rid of the non-alphabetic characters, I used the re library.

After that, as a simple approach, I've iterated through the data and counted the occurances. In the end, the histogram is displayed via matplotlib.

````
import re

dict = {}

regex = re.compile('[^a-zA-Z]')
#First parameter is the replacement, second parameter is your input string

raw_text = regex.sub('', raw_text)


for i in raw_text:
  num = dict.get(i)

  if num:
    dict[i]=num+1
  
  else:
    dict[i]=1



print(dict)

plt.bar(list(dict.keys()), dict.values(),color='g')
plt.show()
````
